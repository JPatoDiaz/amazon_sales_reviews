```python
#Conectar con base de datos externos
import pandas as pd
from google.colab import auth
from google.cloud import bigquery
from IPython.display import display

# Autenticaci√≥n con Google Cloud
auth.authenticate_user()

# Configurar el cliente de BigQuery
project_id = "datalab-414617"
client = bigquery.Client(project=project_id)

# Query para obtener datos de BigQuery - Tabla Amazon Consolidate
query_amazon_consolidate = """
SELECT *
FROM `amazon.consolidate`
"""

# Ejecutar la consulta y almacenar el resultado en un DataFrame de pandas
df_amazon_consolidate = client.query(query_amazon_consolidate).to_dataframe()

# Mostrar todo el DataFrame en formato de tabla
display(df_amazon_consolidate)
```

```python
import pandas as pd
from textblob import TextBlob

# Suponiendo que df_amazon_consolidate es tu DataFrame
data = df_amazon_consolidate.copy()

# Definir una funci√≥n para analizar el sentimiento de cada rese√±a utilizando TextBlob
def analizar_sentimiento(texto):
    analysis = TextBlob(str(texto))
    if analysis.sentiment.polarity > 0:
        return 'positiva'
    elif analysis.sentiment.polarity < 0:
        return 'negativa'
    else:
        return 'neutral'

# Aplicar esta funci√≥n a la columna que contiene las rese√±as y crear una nueva columna para almacenar los resultados del an√°lisis de sentimiento
data['sentimiento'] = data['review_content'].apply(analizar_sentimiento)

# Mostrar el DataFrame con la nueva columna de sentimiento
display(data)

```

```python
# Contar las ocurrencias de cada categor√≠a en la columna 'sentimiento'
conteo_sentimientos = data['sentimiento'].value_counts()

# Calcular el porcentaje de cada categor√≠a
porcentaje_positivas = (conteo_sentimientos['positiva'] / len(data)) * 100
porcentaje_negativas = (conteo_sentimientos['negativa'] / len(data)) * 100
porcentaje_neutras = (conteo_sentimientos['neutral'] / len(data)) * 100

# Mostrar los resultados
print("Porcentaje de rese√±as positivas: {:.2f}%".format(porcentaje_positivas))
print("Porcentaje de rese√±as negativas: {:.2f}%".format(porcentaje_negativas))
print("Porcentaje de rese√±as neutras: {:.2f}%".format(porcentaje_neutras))
```

```python
import pandas as pd

# Suponiendo que data es tu DataFrame
# Calcula los cuartiles
cuartiles = data['discount_percentage'].quantile([0.25, 0.5, 0.75])

# Definir una funci√≥n para clasificar los cuartiles en 'bajo', 'medio' y 'alto'
def clasificar_cuartiles(valor):
    if valor <= cuartiles[0.25]:
        return 'bajo'
    elif valor <= cuartiles[0.5]:
        return 'medio'
    else:
        return 'alto'

# Aplicar la funci√≥n a la columna 'discount_percentage' para crear una nueva columna 'clasificacion_cuartil'
data['clasificacion_cuartil'] = data['discount_percentage'].apply(clasificar_cuartiles)

# Mostrar el DataFrame con la nueva columna de clasificaci√≥n de cuartiles
display(data)

```

```python
# Suponiendo que 'data' es tu DataFrame
variables = data.columns

# Mostrar las variables
print("Variables en la base de datos:")
for variable in variables:
    print(variable)
```

```python
import pandas as pd

# Suponiendo que 'data' es tu DataFrame que contiene los datos
# Si es necesario, aseg√∫rate de tener las columnas en formato num√©rico para calcular las correlaciones

# 1. Correlaci√≥n entre las calificaciones y el precio del producto
correlacion_precio_rating = data[['discounted_price', 'actual_price', 'rating']].corr()

# 2. Correlaci√≥n entre el rating y el sentimiento
correlacion_rating_sentimiento = data[['rating', 'sentimiento']].corr()

# 3. Correlaci√≥n entre la cantidad de calificaciones y el rating
correlacion_rating_count_rating = data[['rating_count', 'rating']].corr()

# 4. Correlaci√≥n entre la categor√≠a del producto y el rating (considerando solo la columna 'category' en este ejemplo)
correlacion_categoria_rating = data[['category', 'rating']].corr()

# 5. Correlaciones entre variables num√©ricas
correlaciones_numericas = data[['discounted_price', 'actual_price', 'rating_count', 'rating']].corr()

# Mostrar las correlaciones
print("Correlaci√≥n entre las calificaciones y el precio del producto:")
print(correlacion_precio_rating)
print("\nCorrelaci√≥n entre el rating y el sentimiento:")
print(correlacion_rating_sentimiento)
print("\nCorrelaci√≥n entre la cantidad de calificaciones y el rating:")
print(correlacion_rating_count_rating)
print("\nCorrelaci√≥n entre la categor√≠a del producto y el rating:")
print(correlacion_categoria_rating)
print("\nCorrelaciones entre variables num√©ricas:")
print(correlaciones_numericas)
```
